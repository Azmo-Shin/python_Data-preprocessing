```python
import pandas as pd

가점 = pd.read_csv('C:/Users/user/Desktop/workSpace/분석/표준분석모델/전기차_교육/건물가점_전처리.csv', engine = 'python')
건물 = pd.read_csv('C:/Users/user/Desktop/workSpace/분석/표준분석모델/전기차_교육/정제데이터/지자체_건물_전처리.csv', engine = 'python')

건물.rename(columns = {'BDTYP_CD' : '건물용도코드', 'BUL_MAN_NO' : '건물일련번호', 'GRO_FLO_CO' : '지상층수'}, inplace = True)

display(건물.head(2))
display(가점.head(2))
```

```python
merged = 건물.merge(가점, on = '건물용도코드')
merged = merged[['건물용도코드', '건물종류', '건물일련번호', '가점척도', '경도', '위도']]
merged.head()
```

```python
가점개수 = len(merged['가점척도'].unique())
bd = [None] * 가점개수
# unipue를 사용해 가점척도 칼럼의 중복행을 제거한 뒤 그 개수만큼 빈 리스트를 생성해 bd로 지정

for i in range(가점개수):
    bd[i] = merged[merged['가점척도']==i]    
    bd[i].to_csv(f'건물{i}.csv', index=False, sep=',', encoding='cp949')    
    display(bd[i].iloc[[0]])    # iloc는 행의 넘버를 뜻함
    
for i in range(가점개수):
    print (f"{i}점 건물수:",len(bd[i]))
```
